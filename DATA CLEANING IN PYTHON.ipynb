```python
# Import libraries
import pandas as pd
```

```python
# Load dataset
orders =pd.read_csv('orders.csv')
```

```python
# View first 10 row in the dataset
orders.head(10)
```

```python
# Check how many values is in the ship mode column
orders['Ship Mode'].unique()
```

```python
# Handle the 'not avilable' and 'unknown' values in ship mode column and treat it as 'null'
    #Handle null values in the dataset#
df=pd.read_csv('orders.csv',na_values=['Not Available', 'unknown'])
```

```python
# Check the ship mode unique values again to confirm your action
df['Ship Mode'].unique()
```

```python
# Change column names to lower case
df.columns=df.columns.str.lower()
```

```python
# Add '_' to the cloumn names
df.columns=df.columns.str.replace(' ','_')
```

```python
df.head()
```

```python
# Create a new column call 'Discount' by calculating the list price and profit
df['discount']=df['list_price']*df['discount_percent']*0.01
```

```python
df.head()
```

```python
# Create a new column call 'sales price ' by calculating the list price and discount
df['sales_price']=df['list_price']-df['discount']
df.head()
```

```python
# Create a new column call 'profit ' by calculating the cost price and sales price
df['profit']=df['sales_price']-df['cost_price']
df.head()
```

```python
# Convert order date column dtype from object to date
df['order_date']=pd.to_datetime(df['order_date'],format="%Y-%m-%d")
```

```python
# Delete cost price, list price, discount percent columns
df.drop(columns=['list_price','cost_price','discount_percent'],inplace=True)
```
